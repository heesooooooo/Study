```python
import numpy as np
```

```python
#1. np.where 함수를 사용하여 평균이 60점 이상이면 pass, 아니면 fail 파생변수 만들기
# 파생변수명 result
exam['result'] = np.where(exam['평균']>=60, 'pass', 'fail')
exam
```

```python
#2. apply 함수 사용하여 평균이 80이상이면 A등급, 70이상이면 B등급, 60이상이면 C등급, 
    # 그 외는 F등급이 주어지는 파생변수 만들기
# 파생변수명 grade

def makeGrade(row) :
    mean = row['평균']
    
    if mean >= 80:
        return 'A'
    elif mean >= 70:
        return 'B'
    elif mean >= 60:
        return 'C'
    else:
        return 'F'
```

```python
exam['grade'] = exam.apply(makeGrade ,axis=1)
exam
```

```python
#3. nclass별 pass와 fail 학생수 확인하기
exam.groupby(['nclass','result']).agg(result_cnt = ('result','count'))
```

```python
#3-1. 번외) pass가 먼저 나오게 바꾸기 : sort_index, sort_value
exam.groupby(['nclass','result']).agg(result_cnt = ('result','count')).sort_index(ascending=False).sort_values(by = 'nclass')
```

```python
#4 nclass별 등급(A~F) 학생 수 확인
exam.groupby(['nclass','grade']).agg(grade_cnt = ('grade','count'))
```

# 데이터 병합

- merge : 컬럼 추가
- concat : 데이터 추가

```python
# merge는 가로로 병합
# 기준이 되는 컬럼을 잡고 새로운 컬럼으로 생성되어 가로로 병합 (옆에 붙는 거..열 추가)
# merge(데이터1, 데이터2, how='left/right', on='컬럼명')
pd.merge(exam, teacher, how='left', on='nclass')
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5],'test':[60,80,70,90,85]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'test':[70,83,65,95,80]})

display(test1)
display(test2)
```

```python
# concat 세로로 병합 pd.concat([])  : 밑에 추가..
pd.concat([test1,test2])
```

```python
# 인덱스를 새롭게 부여하여 증가
# pd.concat([], ignore_index = True)
pd.concat([test1,test2], ignore_index = True)
```

```python
test1 = pd.DataFrame({'id':[1,2,3,4,5],'mid':[60,80,70,90,85]})
test2 = pd.DataFrame({'id':[1,2,3,4,5],'final':[70,83,65,95,80]})

pd.concat([test1,test2])
```

# 결측치
- 누락된 값, 비어있는 값
- 수집 과정에서 오류 발생하거나 포함되어 있지 않은 값 가지고 올 때 
- 데이터의 오류 -> 분석 결과의 왜곡이 발생
- 타이타닉 데이터에 carbin 비어있는 데이터
<br>

- 결측치 해결하는 방법 2가지
    - 1. 채워준다 : 결측치 개수 적을 때 -> 평균값이나 중간값을 채워준다.
    - 2. 버린다 : 결측 개수가 너무 많을 때 -> 컬럼자체를 안 씀
               : 결측치가 적을 때  
               <br>
               
# 이상치
- 정상범위에서 크게 벗어난 값
- 논리적으로 존재할 수 없는 값
- 타이타닉에서 나이 데이터 0~2
<br>

- 이상치 해결하는 방법 2가지
    - 1. 바꿔준다
    - 2. 버린다
