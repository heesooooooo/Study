#### 용어 이해
- 말뭉치(corpus) : 텍스트마이닝, 자연어처리 분야에서 학습을 위해 사용하는 데이터셋 
- 문서(document) : 말뭉치에서 각 샘플을 지칭하는 단어
- TF(Term Frequency) : 하나의 문서(document)에서 개별 단어들이 등장하는 빈도 수
- DF(Document Frequency) : 하나의 단어(토큰)가 전체 말뭉치에서 등장하는 문서(document) 수
